```python

categories = {}
categories_lst = []

for i in dataset:
    if i != None:
        if i["business_id"] != None and i["categories"] != None and "Restaurants" in i["categories"].split(","):
            categories[f'{i["business_id"]},{i["name"]}'] = i["categories"]
            for x in i["categories"].split(","):
                categories_lst.append(x.strip())


                

                    
    
```

```python
# [
#     "Adult Education",
#     "Amateur Sports Teams",
#     "Acupuncture",
#     "Advertising",
#     "Airport Lounges",
#     "",
#     "",
#     ""
# ]
def search_categories():
    y = input()
    for i,v in categories.items():
        for x in v.split(","):
            if y == x.strip():
                z = i.split(",")
                print(z[0],"\n",z[1],"\n", v,"\n")
                with open("test.txt","w") as file:
                    file.write()

search_categories()
```
