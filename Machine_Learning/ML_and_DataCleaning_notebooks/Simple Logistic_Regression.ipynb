## Simple Logistic Regression

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
yelp_data= pd.read_csv("../../Dataset/Cleaned/cleaned_table_2.csv")
```

```python
yelp_data.head()
```

```python
yelp_data.columns
```

```python
# Creating a new empty column is_successful at the end of the Dataset
yelp_data["is_successful"] = ""
yelp_data
```

```python
## Criteria for is_successful
yelp_data.loc[yelp_data["stars"] >= 4.0, "is_successful"] = "yes"
yelp_data.loc[yelp_data["stars"] < 4.0, "is_successful"] = "no"
```

```python
yelp_data.head()
```

```python
# LabelEncoder to turn the labels in the is_successful column to numbers.
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
model_df = yelp_data.copy()
model_df ['is_successful'] = le.fit_transform(model_df['is_successful'])
model_df.head()
```

```python
# Excluding unnecessary columns in the model
x_cols = [i for i in model_df.columns if i not in ('address', 'city', 'state',
                                             'postal_code', 'latitude', 'longitude', 'business_id',
                                             'name','is_successful')]
```

```python
X = model_df[x_cols]
y = model_df['is_successful']
```

```python
# Splitting our data into testing and training sets.
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state=1, stratify=y)
```

```python
# Number of rows and columns in X_train.shape
X_train.shape
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(solver='lbfgs', max_iter=2000,random_state=1)
classifier.fit(X_train, y_train)
```

```python
# Creating predictions for y-values using the X_test set
y_pred = classifier.predict(X_test)
pd.DataFrame({"Prediction": y_pred, "Actual": y_test})
```

```python
# Balanced Accuracy Score
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```
