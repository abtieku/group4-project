```python

categories = {}
categories_lst = []

for i in dataset:
    if i != None:
        if i["business_id"] != None and i["categories"] != None and "Restaurants" in i["categories"].split(","):
            categories[f'{i["business_id"]},{i["name"]}'] = i["categories"]
            for x in i["categories"].split(","):
                categories_lst.append(x.strip())

categories
                

                    
    
```

```python
# [
#     "Adult Education",
#     "Amateur Sports Teams",
#     "Acupuncture",
#     "Advertising",
#     "Airport Lounges",
#     "",
#     "",
#     ""
# ]
def search_categories():
    y = input()
    for i,v in categories.items():
        for x in v.split(","):
            if y == x.strip():
                z = i.split(",")
                print(z[0],"\n",z[1],"\n", v,"\n")

search_categories()
```

```python
new_dict = dict(zip(categories_lst,[0 for i in range(len(categories_lst))]))

for i in categories_lst:
    if i in new_dict.keys():
        new_dict[i] += 1

len(set(categories_lst))
```

```python
data = {}
for key,value in new_dict.items():
    data[key] = [value]
```

```python
business_categories_count = pandas.DataFrame(data, index = ["count"])
```

```python
business_categories_count = business_categories_count.T
business_categories_count.head(50)
```
